**Description:**

- Load one of the Argo index supported by Argopy,
- Search the index,
- Read properties of an index,
- Export the index,
- Floats iterator for more advanced processing.

## Load an Argo index of files

| Index file                                | Shortcut |
|-------------------------------------------|----------|

| argo_synthetic-profile_index.txt          | bgc-s    |
| ar_index_global_meta.txt                  | meta     |

## Searching the index

If you loaded an index, there is a good chance that you will search it with some filters.

An [ArgoIndex (↗)](https://argopy.readthedocs.io/en/v1.3.1/generated/argopy.ArgoIndex.html#argopy.ArgoIndex) instance comes with a `query` extension providing multiple filters that can be used individualy or combined.

### Single filter

All possible **single** filters are (as examples):
```python
idx.query.wmo(1901393)
idx.query.cyc(1)
idx.query.wmo_cyc(1901393, [1,12])

# Taking an index BOX definition:
box = [-60, -55, 40., 45., '2007-08-01', '2007-09-01']
idx.query.lon(box) # Only lon_min/lon_max used
idx.query.lat(box) # Only lat_min/lat_max used
idx.query.lon_lat(box) # Only lon_min/lon_max/lat_min/lat_max used
idx.query.date(box)    # Only date_min/date_max used
idx.query.box(box)

idx.query.params(['C1PHASE_DOXY', 'DOWNWELLING_PAR'])  # Only for BGC profile index
idx.query.parameter_data_mode({'BBP700': 'D'})  # Only for BGC profile index

idx.query.profiler_type(845)
idx.query.profiler_label('NINJA')
```

Let's work with the full BGC-Argo index, `bgc-b`. We can look for profiles in the Ionian Sea in 2024, for instance:

````python
%%time
idx = ArgoIndex(index_file='bgc-b').load()
idx
````

<br>

which can then be filtered like this:

````python
%%time
idx = idx.query.box([15.5, 23, 35, 39, '2024-01', '2025-01'])
idx
````

<br>

The number of files matching the filter is given by the `N_MATCH` attribute:

````python
idx.N_MATCH
````

````python
for af in idx.iterfloats():
    print(af.WMO, af.open_dataset('meta')['LAUNCH_DATE'].values)
````

````python
ds = af.open_dataset('prof')
ds.argo
````

<br>

A quick geographic look at the result can be obtained with the `plot` extension and its `trajectory` method:

````python
idx.plot.trajectory();
````

#### 🔍 Pro tip

The trajectory plot is customizable. For instance if trajectory of each float clutter the figure, it can be removed with the `traj` argument, and free space around plots can be extended with `padding`:

````python
idx.plot.trajectory(traj=False, padding=2);
````

### Multiple filters

It also possible to combine several single filters together with the `compose` method.

In this scenario, filters composition must be provided by a dictionary as an argument to `compose()`.

Each key of the dictionnary is the name of single filter to compose, and filter arguments are passed as the key/value (as a tupple if more than one argument is required by the filter).

Examples:
```python
idx.query.compose({'box': BOX, 'wmo': WMOs})
idx.query.compose({'box': BOX, 'params': 'DOXY'})
idx.query.compose({'box': BOX, 'params': (['DOXY', 'DOXY2'], {'logical': 'and'})})
idx.query.compose({'params': 'DOXY', 'profiler_label': 'ARVOR'})
```

## Index export

Let's load and search an index for export demonstration purposes:

````python
idx = ArgoIndex(index_file='bgc-s').load()
idx
````

````python
idx.query.compose({'params': 'CHLA', 'profiler_label': 'ARVOR'})
````

### Pandas Dataframe

If the [ArgoIndex (↗)](https://argopy.readthedocs.io/en/v1.3.1/generated/argopy.ArgoIndex.html#argopy.ArgoIndex) does not provides you with the most appropriate manipulation methods, you can still export the index, or search results, as a [Pandas Dataframe (↗)](https://pandas.pydata.org/docs/reference/frame.html) with the `to_dataframe` method like this:

````python
df = idx.to_dataframe()
df.head()
````

#### 🔍 Pro tip

If this is more appropriate for the development of your software or procedure, you can restrict export to the first nrows of the index with the `nrows` argument:

````python
idx.to_dataframe(nrows=2)
````

#### 🔍 Pro tip

By default the `to_dataframe()` method will export the search results of your possible query.

If you still want to export the full index, you can use the `index` argument:

````python
df = idx.to_dataframe(index=True)
df.shape
````

### CSV index file

It is also possible that you would be interested to export on file your search results, as a CSV file following the Argo convention. This can be done with the `to_indexfile()` method:

````python
idx.to_indexfile('MyArgoIndexFile.csv')
````

<br>

This file can then be loaded elsewhere with an [ArgoIndex (↗)](https://argopy.readthedocs.io/en/v1.3.1/generated/argopy.ArgoIndex.html#argopy.ArgoIndex):

````python
idx = ArgoIndex(host='.', index_file='MyArgoIndexFile.csv', convention='bgc-s').load()
idx
````

## Floats iterator for more advanced processing

You can loop through Argo index unique floats as [ArgoFloat (↗)](https://argopy.readthedocs.io/en/v1.3.1/generated/argopy.ArgoFloat.html#argopy.ArgoFloat) objects. 

This can be useful to retrieve information from a float that are not in an index, or for batch processing.

As an example, let's get the deployment dates for floats in a specific region:

````python
# Load the index and select profiles in a given box:
idx = ArgoIndex(index_file='bgc-s').query.box([15.5, 23, 35, 39, '2024-01', '2025-01'])
idx
````

````python
print(f"Found {len(idx.read_wmo())} floats / {idx.N_MATCH} profiles in this selection")
````

<br> 

Now we loop on each of these floats. 

The iterator returns a [ArgoFloat (↗)](https://argopy.readthedocs.io/en/v1.3.1/generated/argopy.ArgoFloat.html#argopy.ArgoFloat) object that is used here to load the metadata netcdf file:

````python
for a_float in idx.iterfloats():
    ds = a_float.open_dataset('meta')
    print(a_float.WMO, ds['LAUNCH_DATE'].data)
````

<br>

It's easy to fill in a Dataframe with these new information:

````python
import pandas as pd

data = []
for a_float in idx.iterfloats():
    ds = a_float.open_dataset('meta')
    data.append({'wmo': a_float.WMO, 'LAUNCH_DATE': ds['LAUNCH_DATE'].data})
data = pd.DataFrame(data)
data
````

## Sampling analysis

All the above methods can be combined to conduct a more precise analysis of the Argo sampling over time or space.

#### 🛟 NOTE
A naive approach to this use-case would be to export the full index as a Pandas DataFrame and to work with it.   
However, such an approach would not necessarily be efficient in terms of performances and more importantly would make users to possibly re-implement filters to manipulate complex columns (eg: data modes).  
By using an ``ArgoIndex`` instance all the way down to a plot, users can easily benefit from advanced search features.

<br>
As a trivial example, let's compute the nb of BGC profiles per year:

````python
idx = ArgoIndex(index_file='bgc-b').load()
idx
````

<br>

Get the time period:

````python
date_range = idx.read_domain()[-2:]
date_range
````

<br>

Than compute the annual sampling of all BGC profiles:

````python
import pandas as pd

sampling = []
for interval in pd.interval_range(pd.to_datetime(date_range).year[0], pd.to_datetime(date_range).year[-1]+1):
    sampling.append({'year': interval.left, 
                     'N_MATCH': idx.query.date([-180, 180, -90, 90, str(interval.left), str(interval.right)]).N_MATCH})
sampling = pd.DataFrame(sampling)
sampling['N_MATCH'].describe()
````

<br>

which is easily plotted:

````python
sampling.plot.bar(x='year', grid=True, figsize=(12,6))
````

<br>

We can make the analysis a little bit more precise, by computing the number of profiles with a specific parameter, for instance `DOXY` and `CHLA`:

````python
sampling = []
for interval in pd.interval_range(pd.to_datetime(date_range).year[0], pd.to_datetime(date_range).year[-1]+1):
    sampling.append({'year': interval.left, 
                     'N_BGC': idx.query.compose({'date': [-180, 180, -90, 90, str(interval.left), str(interval.right)]}).N_MATCH,
                     'N_DOXY': idx.query.compose({'params': 'DOXY', 'date': [-180, 180, -90, 90, str(interval.left), str(interval.right)]}).N_MATCH,
                     'N_CHLA': idx.query.compose({'params': 'CHLA', 'date': [-180, 180, -90, 90, str(interval.left), str(interval.right)]}).N_MATCH,
                    })
sampling = pd.DataFrame(sampling)
sampling.plot.bar(x='year', grid=True, figsize=(12,6))
````
