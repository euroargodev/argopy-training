## Make a profile selection with an ArgoIndex

Let's start with the basic import:

````python
# Shut down some warning messages for reading
import warnings
warnings.filterwarnings("ignore")

# Import the Argopy feature we want to work with in this notebook:
from argopy import ArgoIndex
````

and create an instance with a BGC float:

````python
%%time
idx = ArgoIndex(cache=True)
idx.load()
````

````python
idx.query.box([130, 160, 28, 38, '2010','2011'])
idx
````

## Maps of profile locations

An [ArgoIndex (↗)](https://argopy.readthedocs.io/en/lastest/advanced-tools/stores/argoindex.html) instance comes with rudimentary but usefull plotting features.

You can first look at the profile locations, which by default, will also plot float trajectories:

````python
idx.plot.trajectory(legend=False);
````

#### 🔍 Pro tip

The trajectory plot is customizable. 

For instance if trajectory of each float clutter the figure, it can be removed with the `traj=False` argument, and free space around plots can be extended with `padding`.

You can also control some markers properties (size, edgesize and edgecolor).

````python
fig, ax = idx.plot.trajectory(legend=False, traj=False, padding=2, markersize=12, markeredgesize=.1, markeredgecolor='k', palette='Set2');
ax.set_title(f"My selection from the '{idx.convention_title}' index\nSource:{idx.host}");
````

## Bar plot of index properties

The second plotting method is `bar` to draw bar plots of some index properties.

Properties available for bar plots are: 'date', 'latitude', 'longitude', 'ocean', 'profiler_code', 'institution_code', 'date_update', 'wmo', 'cyc', 'institution', 'dac', 'profiler'.

````python
idx.plot.bar('dac');
````
